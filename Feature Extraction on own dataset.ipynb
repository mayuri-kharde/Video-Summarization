# Feature Extraction using Alexnet

```python
import torchvision.models as models
import torchvision.transforms as transforms
from torch.autograd import Variable
from PIL import Image
import torch
import torch.nn as nn
import os
import numpy as np
import pandas as pd 
import h5py
```

```python
alexnet = models.alexnet(pretrained=True)
```

```python
scaler = transforms.Scale((224, 224))
normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
to_tensor = transforms.ToTensor()
```

```python
listing = os.listdir(r'C:/Users/Sweta/Desktop/Project_Dataset/my_dataset/image_dataset')
path = r'C:/Users/Sweta/Desktop/Project_Dataset/my_dataset/image_dataset/'
final_feature_arr = []
video_ids = []
seg_count_arr = []

for category in listing:
    category_path = path + category
    category_listing = os.listdir(category_path)
    
    for video in category_listing:
        video_path = category_path + '/' + video
        video_listing = os.listdir(video_path)
        seg_count = 0
    
        for seg_folder in video_listing:
            seg_path = video_path + '/' + seg_folder
            feature = torch.zeros([1, 1000], dtype=torch.float)
            seg_folder_listing = os.listdir(seg_path)       
            count = 0

            for file in seg_folder_listing:
                file_path = seg_path + '/' + file
                img = Image.open(file_path)
                t_img = Variable(normalize(to_tensor(scaler(img))).unsqueeze(0))
                t = alexnet(t_img)
                feature = torch.add(feature,t) 
                count += 1

            feature = feature / count  
            feature = feature.detach().numpy()
            print(count)

            final_feature_arr.append(feature)  
            seg_count += 1

        video_ids.append(video)
        seg_count_arr.append(seg_count) 
        print(seg_count)

final_feature_arr = np.asarray(final_feature_arr)

with h5py.File('spatial_features_own.hdf5', 'w') as f:
    f.create_dataset('final_feature_arr', data=final_feature_arr)
    
with h5py.File('seg_count_arr_own.hdf5', 'w') as f:
    f.create_dataset('seg_count_arr', data=seg_count_arr)
```

# Feature Extraction using C3D

```python
import skvideo.io
from c3d import C3D
from sports1M_utils import preprocess_input, decode_predictions
import cv2
from keras.models import Model
```

```python
model = C3D(weights='sports1M')
model = Model(inputs=model.input, outputs=model.get_layer('fc6').output)
```

```python
seg_vid_path = 'C:/Users/Sweta/Desktop/Project_Dataset/my_dataset/segmented_videos'
listing = os.listdir(r'C:/Users/Sweta/Desktop/Project_Dataset/my_dataset/segmented_videos')
video_feature_arr = []
temporal_video_ids = []

for category in listing:
    category_path = seg_vid_path + '/' + category
    category_listing = os.listdir(category_path)

    for seg_folder in category_listing:
        seg_video_path = category_path + "/" + seg_folder
        seg_folder_listing = os.listdir(seg_video_path)

        features = np.zeros((1, 4096))

        for segments in seg_folder_listing:
            seg = seg_video_path + "/" + segments

            cap = cv2.VideoCapture(seg)
            fp = cap.get(cv2.CAP_PROP_FPS)    
            frameCount = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            duration = frameCount/fp

            vid = skvideo.io.vread(seg)
            
            for i in range(int(np.floor(duration))):
                video = vid[(int(fp)*i) : (int(fp)*(i+1)+16)]
                x = preprocess_input(video)
                f = model.predict(x)
                features += f
                
            features = features / (i+1)    
            video_feature_arr.append(features)
            print(segments)

        temporal_video_ids.append(seg_folder)
        print(temporal_video_ids)

with h5py.File('temporal_features_own.hdf5', 'w') as f:
    f.create_dataset('video_feature_arr', data=video_feature_arr)
```
